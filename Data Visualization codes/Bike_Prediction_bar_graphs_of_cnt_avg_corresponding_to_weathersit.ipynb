```python
def avg(data,weather_sit):
    cnt=[]
    for i in range(len(data)):
        if data['weathersit'][i]==weather_sit:
            cnt.append(data["cnt"][i])
    return sum(cnt)/len(cnt)
first_quarter=[]
second_quarter=[]
third_quarter=[]
fourth_quarter=[]
first_avg_cnt1=avg(first,1)
first_avg_cnt2=avg(first,2)
first_avg_cnt3=avg(first,3)
first_quarter.append(first_avg_cnt1)
first_quarter.append(first_avg_cnt2)
weather_sit_1=[]
weather_sit_1.append(first_avg_cnt1)
first_quarter.append(first_avg_cnt3)
second_avg_cnt1=avg(second,1)
second_avg_cnt2=avg(second,2)
second_avg_cnt3=avg(second,3)
second_quarter.append(second_avg_cnt1)
second_quarter.append(second_avg_cnt2)
second_quarter.append(second_avg_cnt3)
third_avg_cnt1=avg(third,1)
third_avg_cnt2=avg(third,2)
third_avg_cnt3=avg(third,3)
third_quarter.append(third_avg_cnt1)
third_quarter.append(third_avg_cnt2)
third_quarter.append(third_avg_cnt3)
fourth_avg_cnt1=avg(fourth,1)
fourth_avg_cnt2=avg(fourth,2)
fourth_avg_cnt3=avg(fourth,3)
fourth_quarter.append(fourth_avg_cnt1)
fourth_quarter.append(fourth_avg_cnt2)
fourth_quarter.append(fourth_avg_cnt3)
weather_sit_1.append(second_avg_cnt1)
weather_sit_1.append(third_avg_cnt1)
weather_sit_1.append(fourth_avg_cnt1)
weather_sit_2=[]
weather_sit_2.append(first_avg_cnt2)
weather_sit_2.append(second_avg_cnt2)
weather_sit_2.append(third_avg_cnt2)
weather_sit_2.append(fourth_avg_cnt2)
weather_sit_3=[]
weather_sit_3.append(first_avg_cnt3)
weather_sit_3.append(second_avg_cnt3)
weather_sit_3.append(third_avg_cnt3)
weather_sit_3.append(fourth_avg_cnt3)
```

```python
def plot_bar(quarter,label,quarter_no):
    index = np.arange(len(label))
    plt.bar(index,quarter,align='center',label=label,alpha=0.5)
    plt.ylabel('average', fontsize=15)
    plt.xlabel('weathersit', fontsize=15)
    plt.xticks(index, label, fontsize=15)
    #plt.legend()
    plt.title(quarter_no+' quarter average of cnt corresponding to wearthersit for the year 2011')
    plt.show()
plot_bar(first_quarter,labels,"1st")
plot_bar(second_quarter,labels,"2nd")
plot_bar(third_quarter,labels,"3rd")
plot_bar(fourth_quarter,labels,"4th")
```
